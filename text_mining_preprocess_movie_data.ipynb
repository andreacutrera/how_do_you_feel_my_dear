```python
# variables to select our film and the main characters

movie_name = "the godfather"     # title of the film
movie_id = "m203"                # identifier of the film

tom_hagen = "HAGEN"              # the most loyal to the family - the counselor of don corleone - the lawyer
don_corleone = "DON CORLEONE"    # the main character, the godfather
michael = "MICHAEL"              # the quiet son of don corleone, the successor of the family
clemenza = "CLEMENZA"            # one of the main henchmans 
sonny = "SONNY"                  # the most exhuberant son of the godfather which would be killed by rival family

tom_hagen_id = "u3102"
don_corleone_id = "u3098"
michael_id = "u3105"
clemenza_id = "u3096"
sonny_id = "u3112"
```

```python
characters = speakers[speakers['meta.movie_name'] == movie_name]
characters
```

```python
characters['meta.character_name'].values
```

```python
characters[characters.index == tom_hagen_id]
```

```python
conversations = corpus.get_conversations_dataframe()
conversations.head()
```

```python
conversations = conversations[conversations['meta.movie_name'] == movie_name]
```

```python
conversations
```

```python
conversations['meta.genre'].unique()
```

```python
conversations.index # what we need to filter out the utterances of the film selected
```

```python
utterances = corpus.get_utterances_dataframe()
utterances.head()
```

```python
utterances = utterances[(utterances['meta.movie_id'] == movie_id) &
                        ((utterances['speaker'] == tom_hagen_id) |
                         (utterances['speaker'] == don_corleone_id) |
                         (utterances['speaker'] == michael_id) |
                         (utterances['speaker'] == clemenza_id) |
                         (utterances['speaker'] == sonny_id))]
utterances
```

```python
utterances['speaker'].unique()
```

```python
print(tom_hagen_id, don_corleone_id, michael_id, clemenza_id, sonny_id)
```

```python
lengths = []
for i in range(len(utterances)):
    lengths.append(len(utterances['text'][i].split()))
    
lengths
```

```python
# the longest one, all the others have less than 50 words
utterances['text'][9]
```

```python
code_to_names = {tom_hagen_id: tom_hagen,
                 don_corleone_id: don_corleone,
                 michael_id: michael,
                 clemenza_id: clemenza,
                 sonny_id: sonny}
code_to_names
```

```python
utterances["character"] = utterances["speaker"].map(code_to_names)
utterances
```

```python
utterances = utterances[['text', 'character', 'conversation_id', 'reply_to']]
utterances
```

```python
utterances.to_csv("the_godfather_utterances.csv")
```
